```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path
from types import SimpleNamespace

calc_dir = Path.cwd().joinpath('../calcs')
```

```python
def parse_path(p):
    atom, code = p.stem.split('_')
    return atom, bool(int(code[0])), bool(int(code[1])), bool(int(code[2]))
```

```python
table = []
for output in calc_dir.glob('*.json'):
    atom, has_nuc, has_coul, has_xc = parse_path(output)
    with output.open() as f:
        j = json.load(f, object_hook=lambda x: SimpleNamespace(**x))
        if not j.output.success: print('Job failed: ', output.stem)
        
    e = j.output.properties.scf_energy.E_tot
    table.append((atom, has_nuc, has_coul, has_xc, e))
    
df = pd.DataFrame(table, columns=['Element', 'include_nuclear', 'include_coulomb', 'include_xc', 'Total_Energy']).sort_values(by=['Element', 'include_nuclear', 'include_coulomb', 'include_xc']).reset_index(drop=True)
```
